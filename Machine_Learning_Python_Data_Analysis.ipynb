```python
import matplotlib.pyplot as plt

plt.title('Scree Plot')
plt.plot(range(1, 23), pca.explained_variance_ratio_, 'o-')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xticks(np.arange(0, 24, 1))
plt.yticks(np.arange(0, 0.5, 0.025))
plt.grid(True)
plt.text(2, -0.12, 'Figure 1. Principal Component Scree Plot', c = '#9ACD32', fontsize = 'x-large')

plt.show()
```

```python
for i in range(2):
    print(f'{(round(pca.explained_variance_ratio_[i], 3)*100):.1f}%')
```

```python
#KMEANS CLUSTERING
#Method 1:

from sklearn.cluster import KMeans

# Elbow Method to determine optimal K for K-Means Clustering
wcss = []
for i in range(1,10):
    kmeans = KMeans(n_clusters = i, n_init = 'auto')
    kmeans.fit(df2_s)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10), wcss, 'o-')
plt.xlabel('Number of clusters')
plt.ylabel('Within-Cluster Sum of Square')
plt.title('Elbow Method for K-Means Clustering')
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(2*1e6, 6.5*1e6, 0.5*1e6))
plt.grid(True)
plt.text(2.15, 1.1*1e6, 'Figure 3. Elbow Method Plot', c = '#9ACD32', fontsize = 'x-large')

plt.show()
```

```python
#from the Graph of the Elbow method, there is no clear elbow shape show where there is a huge decrease in difference of the gradient between each cluster
#thus, we will be computing the Silhouette score to determine the optimal number of clusters to use.

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

n = int(0.1* len(df2_s)) #Sample size 10% for computing silhouette score as the kmeansfit will be too computationally expensive for me to conduct
k_range = range(2, 10)
silhouette_scores = [[] for _ in range(5)]
for k in k_range:
    for i in range(2024, 2029):
        df2_s_sample = df2_s.sample(n = n, random_state = i)
        kmeans = KMeans(n_clusters = k, n_init = 'auto', random_state = i)
        kmeans.fit(df2_s_sample)
        ss_value = silhouette_score(df2_s_sample, kmeans.labels_)
        silhouette_scores[i-2024].append(ss_value)
        print('Sample:', i, ' | Clusters:', k, ' | Silhouette Score:', ss_value)
```

```python
ssdf = pd.DataFrame(silhouette_scores, columns = ['2', '3', '4', '5', '6', '7', '8', '9'])
ssdf
```

```python
ss_list = []
for col in ssdf.columns:
    mean_ss = ssdf[col].mean()
    ss_list.append(mean_ss)

print(ss_list)
```

```python
plt.plot(range(2,10), ss_list, 'o-')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Silhouette Scores')
plt.title('Silhouette Analysis for K-Means Clustering')
plt.grid(True)
plt.text(3, 0.103, 'Figure 4. Silhouette Method Plot', c = '#9ACD32', fontsize = 'x-large')

plt.show()
```

```python
df2 = pd.read_csv('C://Users/vyndj/ST3189_ML_Coursework/Data/heart_disease_health_indicators_BRFSS2015.csv')
df2.info()
```

```python
x = df2.drop(['HeartDiseaseorAttack'], axis=1) #split into x and y variables
y = df2['HeartDiseaseorAttack']
```

```python
x_categorical = x.drop(['BMI', 'MentHlth', 'PhysHlth'], axis = 1)
x_numerical = x[['BMI', 'MentHlth', 'PhysHlth']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_numerical_s = pd.DataFrame(scaler.fit_transform(x_numerical), columns = x_numerical.columns)

x = x_categorical.join(x_numerical_s)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2024)
```

```python
#K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier

#to determine the optimal K value
test_scores = []
for i in range(1, 16):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(x_train, y_train)
    test_scores.append(knn.score(x_test, y_test))
    print(i, 'done')
```

```python
test_scores
```

```python
import numpy as np
import seaborn as sns

plt.figure(figsize = (12,6))
sns.lineplot(x = range(1,16), y = test_scores)
plt.xticks(np.arange(0, 16, step = 1))
plt.yticks(np.arange(0.86, 0.91, step = 0.01))
plt.grid(True)
plt.title('Test Scores at each K for KNN')
plt.xlabel('Value of K')
plt.ylabel('Test score')
plt.axhline(y = max(test_scores), linestyle = '--', color = 'k')
plt.text(5, 0.852, 'Figure 7. Test Scores at each K', c = '#9ACD32', fontsize = 'x-large')

plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 12)
knn.fit(x_train, y_train)
knn_pred = knn.predict(x_test)
knn.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm_knn = confusion_matrix(y_test, knn_pred)
print(classification_report(y_test, knn_pred, digits = 3))
```

```python
from sklearn import metrics
cm_knnd = metrics.ConfusionMatrixDisplay(confusion_matrix = cm_knn, display_labels = ['No Heart Disease', 'Have Heart Disease'])
cm_knnd.plot()
plt.title('Confusion Matrix for K-Nearest Neighbors')

plt.show()
```

```python
from sklearn.metrics import roc_curve
knn_pred_proba = knn.predict_proba(x_test)[:,1]
fpr_knn, tpr_knn, _ = roc_curve(y_test, knn_pred_proba)

plt.plot(fpr_knn, tpr_knn, linestyle = '--',color = 'blue')
plt.xticks(np.arange(0, 1.1, step = 0.1))
plt.yticks(np.arange(0, 1.1, step = 0.1))
plt.grid(False)
plt.axline((0, 0), (1, 1), linestyle = '--', color = 'black')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for K-Nearest Neighbors')

plt.show()
```

```python
## Linear Regression
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2024) #split into test and train, test size being 20%, and set seed of random_state = 2024

lm=LinearRegression()
lm.fit(x_train, y_train)
lm_pred = lm.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error as mse

print('mse for linear regression:', mse(y_test, lm_pred))
print('Rmse for linear regression:', mse(y_test, lm_pred, squared = False))
print('r2 score for linear regression:', lm.score(x_test,y_test)) #0.6790045635265182
```
